### **Exercise 31**

Consider the following array

[[4.99, 3.49, 9.99],<br>
[1.99, 9.99, 4.99],<br>
[14.99, 2.39, 7.29]]

sort this array:

* by row (ascending)
* by column (ascending)



```python
A = np.array([[4.99, 3.49, 9.99],
              [1.99, 9.99, 4.99],
              [14.99, 2.39, 7.29]])
```

```python
# by column
np.sort(A, axis=0)
```

```python
# by row
np.sort(A, axis=1)
```
