### **Exercise 101**

Solve the following system of equations

$5x - 3y = 21$ <br>
$x - 2y =7$



```python
A = np.array([[5, -3], [1, -2]])
B = np.array([[21], [7]])
X = np.dot(np.linalg.inv(A), B)
x = X[0][0]
y = X[1][0]
print(f'x = {x:.2f}\ny = {y:.2f}')
```

### **Exercise 102**

Solve the following system of equations

$x + y + z = 1$ <br>
$2x + y + 5z = 0$ <br>
$x - y = z$ 




```python
A = np.array([[1, 1, 1], [2, 1, 5], [1, -1, -1]])
B = np.array([[1], [0], [0]])
X = np.dot(np.linalg.inv(A), B)
x = X[0][0]
y = X[1][0]
z = X[2][0]
print(f'x = {x:.2f}\ny = {y:.2f}\nz = {z:.2f}')
```
